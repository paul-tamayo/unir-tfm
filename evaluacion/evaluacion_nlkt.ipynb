```python
from sklearn import metrics

# Constants
H = "HATE"
O = "NO_HATE"

# valores Reales
y_true = [
    O, H, H, H, H, H, H, H, H, H,
    H, H, H, H, H, H, H, H, H, H,
    H, H, H, H, H, H, H, H, H, H,
    H, H, H, H, H, H, H, H, H, H,
    O, H, H, H, H, H, H, H, H, H,
    H, H, H, H, H, H, H, H, H, H,
    H, H, H, O, H, H, O, O, H, H,
    O, H, H, H, H, O, H, H, H, H,
    H, H, H, H, H, H, H, H, H, H,
    H, H, H, H, H, H, H, H, H, H,
    H, H, H, H, H, H, H, H, H, H,
    H, H, O, H, H, O, O, H, H, O
]
# valores Predichos
y_pred = [
    O, H, H, O, H, O, H, H, O, H,
    H, H, H, H, H, H, H, H, H, H,
    H, H, O, H, H, O, O, O, H, H,
    H, H, O, H, H, H, O, H, H, H,
    H, O, O, O, O, O, O, H, H, O,
    H, H, H, O, O, H, H, O, H, H,
    H, H, O, O, H, H, O, O, H, H,
    O, H, O, H, H, O, O, H, H, H,
    O, H, H, O, H, H, H, H, O, H,
    H, H, H, H, H, O, O, O, O, H,
    O, H, O, O, O, H, H, O, H, H,
    H, H, O, H, H, O, H, O, O, O
]

# Mostrar confusion matrix
matrix = metrics.confusion_matrix(y_true, y_pred)

# Mostrar precision and recall, junto con el resto de m√©tricas
print(metrics.classification_report(y_true, y_pred, digits=3))
```

```python
import matplotlib.pyplot as plt
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=matrix, display_labels=[H, O])
cm_display.plot()
plt.show()
```
